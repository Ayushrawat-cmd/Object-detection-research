```python
import torch 
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

# Load Model

```python
model = torch.hub.load('ultralytics/yolov5','yolov5s') #we have used the small version of yolo here
```

```python
model
```

# Make detections with images

```python
 img = "https://ultralytics.com//images/zidane.jpg"
```

```python
results = model(img)
results.print()
```

```python
img_cars = "https://images.indianexpress.com/2021/09/Delhi-rain-2-1.jpg"
```

```python
cars_result = model(img_cars)
cars_result.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(cars_result.render()))
plt.show()
```

# Real time detections 

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    #make detections 
    results = model(frame)
    
    cv2.imshow('YOLO',np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
```
